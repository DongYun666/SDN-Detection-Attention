```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./dataset/CIC_IDS_2017/CIC_IDS_2017.csv',skiprows=1,header=None)
```

```python
# 复制数据集
dataset = data.copy()
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
# 删除[0,1,2,3]列
dataset.drop([0,1,2,3],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
# 删除重复行
dataset.drop_duplicates(inplace=True)
```

```python
# 删除缺失值
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.iloc[:,-1].value_counts()
```

```python
dataset.head()
```

```python
# 数据归一化
dataset.iloc[:,:-8] = np.arctan(dataset.iloc[:,:-8]) * 2 / np.pi
```

```python
dataset.head()
```

```python
dataset[63].value_counts()
```

```python
# 对多类别列 使用labelencoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset[63] = le.fit_transform(dataset[63])
```

```python
dataset.head()
```

```python
type(dataset.iloc[0,0])
```

```python
# 采用train_test_split 划分数据集
from sklearn.model_selection import train_test_split

train,test = train_test_split(dataset,test_size=0.2,random_state=226)
```

```python
type(train.iloc[0,0])
```

```python
# 保存数据集 wei npy文件
np.save('./processdata/CIC_IDS_2017/train.npy',train.values)
np.save('./processdata/CIC_IDS_2017/test.npy',test.values)
```
